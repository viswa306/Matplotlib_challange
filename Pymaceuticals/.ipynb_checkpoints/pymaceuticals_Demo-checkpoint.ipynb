```python
# Based on the above data grouped themouseid and find the tumoraverage.

avg_tumor_vol =study_results_df.groupby('Mouse ID').median()["Tumor Volume (mm3)"]
avg_tumor_vol
avg_tumorvol_df= pd.DataFrame(avg_tumor_vol)
weight_avgtumor=avg_tumorvol_df.reset_index()
weight_avgtumor

weight_avgtumor_df=pd.merge(Capomulin_df ,weight_avgtumor,on="Mouse ID",how="outer")
weight_avgtumor_df
```

```python
 # Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen. 
plt.scatter(weight_avgtumor_df.iloc[:,4],weight_avgtumor_df.iloc[:,5])
plt.title('Treatment for 4 regimens')
plt.xlabel(' Mouse Weight (g)')
plt.ylabel('Mouse Tumor Volume (mm3)')
plt.show()
```

```python
#Calculate the correlation coefficient and linear regression model
Weight= weight_avgtumor_df.iloc[:,4]
TumorVolume = weight_avgtumor_df.iloc[:,5]
correlation = st.pearsonr(Weight,TumorVolume)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
x_values =  weight_avgtumor_df['Weight (g)']
y_values = weight_avgtumor_df['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(0,50),fontsize=15,color="red")
plt.xlabel('Normalized Blood Pressure')
plt.ylabel('1Y_Disease_Progress')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
